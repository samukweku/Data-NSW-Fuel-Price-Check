```python
import pandas as pd
import numpy as np
import glob
```

```python
#filter for only filenames that have digits before the .csv
fuel_datasets_dir='/home/sam/Github Folder/FuelCheck/*[0-9]*.csv'
```

```python
#generate a list of filenames
l=[l for l in glob.glob(fuel_datasets_dir)]
```

```python
#run a for loop to get all the files into pandas
fuel_list=[]
for i in l:
    fuel_list.append(pd.read_csv(i))
```

```python
#not sure if i will need this, just stripped the month and year, to add 
#to the fuel_table as an index
j=[j.split('/')[-1].lower().rstrip('.csv') for j in l]
```

```python
#concatenate all the files into one table
fuel_table=pd.concat(fuel_list, sort=True)
```

```python
#an irrelevant column, as such i dropped it
fuel_table = fuel_table.drop(columns='Unnamed: 0')
```

```python
# 'merge' done
fuel_table['FuelCode']=np.where(fuel_table['FuelCode'].isnull(),fuel_table['FuelType'],fuel_table['FuelCode'])
```

```python
#fueltype column removed
fuel_table.drop('FuelType',axis=1,inplace=True)
```
